```python
"""
Code to test out how .ast (ascii stl) files work
"""

import os
import numpy as np
import plotly.graph_objects as go


def print_file(file_path):
    """
    Prints the contents of a text file
    :param file_path: location of file ending in file name and extension
    :return:
    """
    file = open(file_path, 'r')
    print(file.read())
    file.close()


def rm_newline(string):
    """

    :param string:
    :return:
    """
    if string.__contains__('\n'):
        return string.replace('\n', '')
    else:
        return string


def read_object3d(file_path):
    """

    :param file_path:
    :return:
    """
    opened_file = open(file_path, 'r')
    begin_facet_string = 'facet normal'
    endof_facet_string = 'endfacet'
    begin_vertex_string = 'outer loop'
    endof_vertex_string = 'endloop'
    object_tuple = []
    triangle_vertices = []
    triangle_count = 0
    for line in opened_file:
        # Remove '\n' from end of string
        line = rm_newline(line)
        if line.__contains__(begin_facet_string):
            object_tuple.append([])
            # Extract x, y, and z normal coordinates from facet data
            normal_coords = line.split(' ')[4:]
            # Add face name and normal coordinates to faces_tuple
            object_tuple[triangle_count].append('Face ' + str(triangle_count))
            object_tuple[triangle_count].append(normal_coords)
        if line.__contains__(begin_vertex_string):
            triangle_vertices = []
        if line.__contains__('vertex'):
            vertex_coords = line.split(' ')[7:]
            triangle_vertices.append(vertex_coords)
        if line.__contains__(endof_vertex_string):
            object_tuple[triangle_count].append(triangle_vertices)
        if line.__contains__(endof_facet_string):
            triangle_count += 1
    opened_file.close()
    return object_tuple


def print_object3d(object_tuple):
    """

    :param object_tuple: list of all data extracted from .ast file
    :return:
    """
    print('Number of faces: {}'.format(len(object_tuple)))
    for simplex in object_tuple:
        print(simplex[0] + ':')
        print('   Normal: \n'
              '      x: {}\n'
              '      y: {}\n'
              '      z: {}'.format(simplex[1][0], simplex[1][1], simplex[1][2]))
        print('   Vertices:\n'
              '      vertex 1:\n'
              '         x: {}\n'
              '         y: {}\n'
              '         z: {}\n'
              '      vertex 2:\n'
              '         x: {}\n'
              '         y: {}\n'
              '         z: {}\n'
              '      vertex 3:\n'
              '         x: {}\n'
              '         y: {}\n'
              '         z: {}'.format(simplex[2][0][0], simplex[2][0][1], simplex[2][0][2],
                                      simplex[2][1][0], simplex[2][1][1], simplex[2][1][2],
                                      simplex[2][2][0], simplex[2][2][1], simplex[2][2][2]))


def list_all_vertices(object_tuple, test_code=False):
    """
    Creates list of all vertices in 3d object
    :param object_tuple: list of all data extracted from .ast file
    :param test_code: boolean variable to execute print statemts for testing
    :return:
    """
    # Print vertices for each face
    coord_string = ['x', 'y', 'z']
    all_vertices_list = []
    if test_code:
        print('\nPrinting vertices for each face')
    for face_count in range(len(object_tuple)):
        if test_code:
            print('Vertices for face {}'.format(face_count))
        for vertex_count in range(len(object_tuple[face_count][2])):
            vertex_list = []
            if test_code:
                print('   Vertex {}:'.format(vertex_count))
            for coord in range(3):
                if test_code:
                    print('      {}: {}'.format(coord_string[coord], object_tuple[face_count][2][vertex_count][coord]))
                vertex_list.append(object_tuple[face_count][2][vertex_count][coord])
            all_vertices_list.append(vertex_list)
    return all_vertices_list


def unique_points(object_tuple):
    """
    Finds unique vertices/points in the 3d object
    :param object_tuple: list of all data extracted from .ast file
    :return: list of tuples with unique points from .ast file
    """
    all_vertices_list = list_all_vertices(object_tuple)
    unique_list = []
    for a in range(len(all_vertices_list)):
        for b in range(len(all_vertices_list)):
            if all_vertices_list[a] == all_vertices_list[b] and not unique_list.__contains__(all_vertices_list[b]):
                unique_list.append(all_vertices_list[b])
    return unique_list
```

```python
# Find working directory of python file and .ast files
pwd = os.getcwd()+'/stl_files/
print(pwd)

# Create empty list for storing filenames
ast_list = []

# Run through list and add files with .ast extension to ast_list
for list_item in os.listdir(pwd):
    if list_item.__contains__('.ast'):
        ast_list.append(list_item)

# Prompt user for filename
for face in range(len(ast_list)):
    print(str(face) + ' ' + ast_list[face])
ast_index = int(input('Choose .ast file by entering the corresponding number:\n'))
chosen_file = ast_list[ast_index]
chosen_file_path = pwd + '/' + chosen_file
print('chosen .ast file: {}'.format(chosen_file))
print('chosen .ast filepath:\n{}\n'.format(chosen_file_path))
print('Printing file as-is:')
print_file(chosen_file_path)

# Read in data from .ast file
print('Printing file after storing data to tuple:')
object3d = read_object3d(chosen_file_path)
print_object3d(object3d)


```
