```python
import os
import time
import networkx as nx
from sympy import Matrix
import plotly.graph_objects as go
from meshpy import tet
import shutil
import gudhi
from tqdm import tqdm
import matplotlib.pyplot as plt
from meshpy import tet
```

```python
import numpy as np
np.random.seed(2023)
```

```python
def create_tetrahedrons(file, path, switch='pq/10a50V',
                        timer=False, verbose=False):
    """
    Uses meshpy to create a list of tetrahedrons in a
    mesh generated by meshpy
    :param file: file name with extension
    :param path: file path with extension
    :param switch: https://wias-berlin.de/software/tetgen/switches.html
    :param timer: boolean variable to count computation time
    :param verbose: prints progress statements
    :return: tetrahedron mesh
    """
    time_start = 0
    if timer:
        time_start = time.time()
    tet_options = tet.Options(switches=switch)
    meshinfo = tet.MeshInfo()
    src_dir = path
    temp_string = ' (temp STL conversion)'
    dst_dir = file[:-4] + temp_string + '.stl'
    shutil.copy(src_dir, dst_dir)
    meshinfo.load_stl(dst_dir)
    tetra_mesh = tet.build(meshinfo, options=tet_options, verbose=verbose)
    os.remove(dst_dir)
    if timer:
        sca.func_timer(time_start)
    return tetra_mesh
```

```python
def create_mesh_array(tetra_mesh,
                      verbose=False):
    """
    Concatenates items in MeshPy mesh in order of
    points, edges, faces, and tetrahedrons
    :param tetra_mesh: Mesh generated by MeshPy
    :param verbose: prints progress statements
    :return: A list matching the formatting of the first indices in
    the list generated by gudhi vietoris-rips complex.
    """
    mesh_array = []

    # Create array of points from tetrahedra
    point_array = []
    for tetra in list(tetra_mesh.elements):
        for point in tetra:
            if not point_array.__contains__(point):
                point_array.append(point)
    point_array = sorted(point_array)
    point_array = [[point - 1] for point in point_array]

    # Create array of faces from tetrahedra
    face_array = []
    for tetra in list(tetra_mesh.elements):
        tetra = sorted(tetra)
        for index in range(4):
            face_array.append([tetra[0 - index] - 1,
                               tetra[1 - index] - 1,
                               tetra[2 - index] - 1])
    face_array = [sorted(face) for face in face_array]

    # Create array of edges from array of faces
    edge_array = []
    for face in face_array:
        for index in range(3):
            edge = sorted([face[0 - index], face[1 - index]])
            if not edge_array.__contains__(edge):
                edge_array.append(edge)
    edge_array = sorted([sorted(edge) for edge in edge_array])

    mesh_array.extend(point_array)
    mesh_array.extend(edge_array)
    mesh_array.extend(face_array)
    mesh_array.extend([sorted([item - 1 for item in array]) for array in list(tetra_mesh.elements)])
    if verbose:
        print('MeshPy Mesh Items:')
        for item in mesh_array:
            print(item)
    return mesh_array

def create_complex(tetra_mesh, complex_type='a', verbose=False):
    """
    Creates a complex using gudhi
    :param tetra_mesh: MeshPy mesh
    :param complex_type: type of complex, r for vietoris-rips, a for alpha
    :param verbose: prints progress statements
    :return: vietoris rips complex, simplex tree from rips complex, rips filtration
    """
    complex_type = complex_type.lower()
    complex_name = ''
    if complex_type == 'a':
        complex_name = 'Alpha'
    elif complex_type == 'r':
        complex_name = 'Vietoris Rips'
    total_mesh = create_mesh_array(tetra_mesh)
    gudhi_complex = None
    gudhi_simplex_tree = None
    if complex_type == 'r':
        print('Creating Rips Complex ...')
        # noinspection PyUnresolvedReferences
        # gudhi_complex = gudhi.RipsComplex(points=tetra_mesh.points, max_edge_length=max_distance)
        gudhi_complex = gudhi.RipsComplex(points=tetra_mesh.points)
        print('Creating Simplex Tree ...')
        gudhi_simplex_tree = gudhi_complex.create_simplex_tree(max_dimension=3)
    elif complex_type == 'a':
        print('Creating Alpha Complex ...')
        # noinspection PyUnresolvedReferences
        gudhi_complex = gudhi.AlphaComplex(points=tetra_mesh.points)
        print('Creating Simplex Tree ...')
        gudhi_simplex_tree = gudhi_complex.create_simplex_tree()

    if verbose:
        print('\n' + complex_name + ' Complex, Dim = {}'.format(gudhi_simplex_tree.dimension()))
        print(
            'vertices:  {}\nsimplices: {}'.format(gudhi_simplex_tree.num_vertices(),
                                                  gudhi_simplex_tree.num_simplices()))

    print('Reassigning filtration values based on existing mesh items ...')
    zero_counter = 0
    for filtered_item in tqdm(gudhi_simplex_tree.get_filtration()):
        if total_mesh.__contains__(filtered_item[0]):
            gudhi_simplex_tree.assign_filtration(filtered_item[0], 0)
            zero_counter += 1
    if verbose:
        print('\n' + str(zero_counter) + ' mesh items reassgined')
        print(complex_name + ' Filtration Items:')
        if zero_counter <= 25:
            print('Printing reassigned filtration values...')
            for item in gudhi_simplex_tree.get_filtration():
                if total_mesh.__contains__(item[0]):
                    print(item)
        print('Printing persitence diagram points...')
        for item in gudhi_simplex_tree.persistence():
            print(item)
    return gudhi_complex, gudhi_simplex_tree
```

```python
chosen_file_path = "cube - triangle hole.ast"
T = create_tetrahedrons(chosen_file_path, 'STL Experiments 4/cube - triangle hole.ast', switch='a5V')
object3d_rips_complex, object3d_simplex_tree = create_complex(T, complex_type='a', verbose=False)
```
